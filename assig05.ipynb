```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv('/content/carscsv.csv')
a
```

```python
#line chart of kilometer and price
x=a['Kilometer']
y=a['Price']
plt.plot(x,y,marker='o')
plt.title('kilometer and price analysis')
plt.xlabel('Kilometer')
plt.ylabel('Price')
plt.show()
```

```python
#company and fuel tank capacity - Bar plot
categories=a['Make']
values=a['Fuel Tank Capacity']
plt.bar(categories,values)
plt.title('company and fuel analysis')
plt.xlabel('fuel tank capacity')
plt.ylabel('Company')
```

```python
#histogram
c=a['Price']
plt.hist(c,bins=20,edgecolor='k')
plt.title('Frequency of Car price')
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.show()
```

```python
#year and kilometer scatter plot
d=a['Year']
e=a['Kilometer']
plt.scatter(d,e)
plt.title('kilometer analysis')
plt.xlabel('year')
plt.ylabel('kilometer')
plt.show()
```

```python
f=a.head(10)
categories=f['Make']
values=f['Year']
plt.pie(values,labels=categories,autopct='%1.1f%%')
plt.title("pie")
plt.show()
```

```python
g=a['Price']
plt.hist(g,density=True,bins=30,alpha=0.5)
plt.xlabel('values')
plt.ylabel('Density')
plt.show()
```
